<p align="justify">As ocorrências mais frequentes de fraude financeira são: cartão perdido ou roubado, cartão não presente, falsificação, roubo de identidade e cartão nao recebido. A seguir, uma breve descrição dos tipos de ocorrências:</p><br>
<center>

| Fraude Financeira            | Descrição (baseado nos registros de reclamação)|
|------------------------------|--------------------------------------------|
| **Cartão perdido ou roubado**| Ocorre quando um criminoso usa um cartão perdido ou roubado para fazer compras ou pagamento fraudulento. fraudulento.|
| **Cartão não presente**      |  É um tipo de golpe em que o criminoso usa as informações do cartão do titular para fazer transações fraudulentas pela Internet e telefone.                              |
| **Falsificação**             | Ocorre quando um criminoso cria uma cartão falso, usando as informações obtidas da tarja magnética do cartão do titular.                                                                    |
| **Roubo de identidade**      | Aquisição e uso indevido de informações de identificação privada de um indivíduo para obter acesso a fundos ou cometer crimes.                                                                     |
| **Cartão nao recebido**      | Fraude que acontece quando um cartão é roubado, enquanto está em trânsito entre o emissor do cartão e o titular do cartão.                                                                     |

###**3.5 Verificando o balanceamento do dataset**

<p align="justify">Quando um conjunto de dados possui uma ou mais classes, como é o caso do dataset "creditcard.csv" que contém transações de cartão de crédito rotuladas como fraudulentas ou legítimas, é importante verificar se as classes estão distribuídas de maneira equilibrada ou desiquilibrada.</p>

<p align="justify">Em outras palavras, a distribuição desequilibrada ocorre quando uma classe possui um número significativamente maior ou menor de exemplos em relação a outra classe. No conjunto de dados "creditcard.csv", se verificarmos que existe uma proporção muito pequena de transações fraudulentas em relação às transações legítimas, vamos confirmar que o dataset está desbalanceado.</p>

<p align="justify">Esse desequilíbrio pode levar a um desempenho insatisfatório do modelo de aprendizado de máquina, pois o modelo pode ser treinado para ser excessivamente sensível à classe majoritária (transações legítimas) e não ser capaz de detectar corretamente a classe minoritária (transações fraudulentas). Isso pode levar a resultados incorretos, como a identificação incorreta de transações legítimas como fraudulentas ou vice-versa.</p>

<p align="justify">Portanto, é importante verificarmos se o conjunto de dados está balanceado antes de treinarmos o modelo de aprendizado de máquina para garantir que o modelo aprenda de forma justa e precisa, a partir dos exemplos de todas as classes.</P>

<p align="justify">O cálculo da correlação é uma técnica quantitativa que permite medir a força e a direção da relação linear entre duas variáveis numéricas. No entanto, a correlação não é capaz de detectar relações não-lineares entre as variáveis, nem é adequada para identificar relações causais.</p>

<p align="justify">A plotagem de gráficos de densidade, por outro lado, é uma técnica visual que permite avaliar a distribuição de uma variável numérica em relação a outra variável numérica ou categórica. Essa técnica pode ser útil para identificar padrões complexos e não-lineares entre as variáveis, bem como para detectar outliers ou valores extremos.</p>

<p align="justify">A seguir, vamos plotar um gráfico de densidade para compararmos as distribuições de cada variável (V1 a V28) em relação às classes (0 e 1), visando a detecção de anomalias.</p>

###**3.7 Etapas do processo de detecção de fraudes em cartões de crédito**

<p align="justify">O processo de detecção de fraudes em cartões de crédito geralmente envolve várias etapas, e a ordem pode variar dependendo da abordagem utilizada. No entanto, em geral, é comum realizarmos o seguinte processo:

<p align="justify">A) Preparação dos dados: inclui etapas como limpeza, transformação, normalização, seleção de recursos e balanceamento de classes.</p>

<p align="justify">B)Treinamento do modelo de machine learning: um modelo de machine learning é treinado usando os dados preparados. Geralmente, são utilizados algoritmos supervisionados, como regressão logística, árvores de decisão, redes neurais, entre outros.</p>

<p align="justify">C) Avaliação do modelo: o modelo é avaliado usando métricas de desempenho, como precisão, recall, F1-score, AUC, entre outras. Essas métricas ajudam a entender o quão bem o modelo está performando na detecção de fraudes.</p>

<p align="justify">D) Detecção de anomalias: algoritmos de detecção de anomalias são usados para identificar transações suspeitas que não se enquadram nos padrões normais do conjunto de dados. Isso pode ajudar a detectar fraudes que não foram identificadas pelo modelo de machine learning.</p>

<p align="justify">A primeira etapa da preparação dos dados, requer uma atenção especial às colunas 'Time' e 'Amount', pois estas variáveis não estão normalizadas. Isso significa dizer que essas colunas não estão em uma escala comum ou padronizada.</p>

<p align="justify">O conjunto de dados de treinamento é usado para ajustar os parâmetros do modelo de aprendizado de máquina, ou seja, para fazer com que o modelo "aprenda" a relação entre as variáveis de entrada e de saída, enquanto que o conjunto de validação é usado para avaliar a qualidade do modelo.

## **6. Avaliando o desempenho do modelo (Matriz de Confusão)**

###**6.1 Matriz de Confusão**

<p align="justify">Para avaliarmos um modelo de machine learning, primeiramente precisamos construir uma matriz de confusão e usarmos as métricas de desempenho para entender o quão bem o modelo está performando na detecção de fraudes. 

<p align="justify">A matriz de confusão é uma tabela que mostra a performance de um modelo de classificação em relação aos dados de teste. A matriz é composta por quatro elementos principais: verdadeiro positivo (TP), falso positivo (FP), verdadeiro negativo (TN) e falso negativo (FN).</p>

<p align="justify">A matriz de confusão é diretamente relacionada às métricas de avaliação de um modelo de classificação, como precisão, recall e área sob a curva (AUC). A partir dos valores de TP, FP, TN e FN, é possível calcular essas métricas.</p>

<p align="justify">O quadrante 1 (TP - True Positive), representa o número de instâncias que foram corretamente classificadas como positivas. O quadrante II (FP - False Positive), representa o número de instâncias que foram incorretamente classificadas como positivas.</p> 

<p align="justify">O quadrante III (FN - False Negative), representa o número de instâncias que foram incorretamente classificadas como negativas. E o quadrante IV (TN - True Negative), representa o número de instâncias que foram corretamente classificadas como negativas, conforme demonstrado abaixo:</p><br>

<center><img alt="aci_universal_payments" width="100%" src="https://blog.roboflow.com/content/images/2022/11/confussion-matrix-2.png"></center>

<p align="center"><font size="2">Fonte: <a href="https://blog.roboflow.com/what-is-a-confusion-matrix/"> https://blog.roboflow.com/JamesGallagher</a></p></font><br>

###**6.2 Comparação com outros modelos**

<p align="justify">Para fins de estudo, já testamos a técnica de "under-sampling" sobre o conjunto de dados. Com o intuito de melhorarmos o desempenho do modelo de machine learning, vamos testar um segundo modelo, chamado de SMOTE (Synthetic Minority Over-sampling Technique), que é uma técnica de "over-sampling" (superamostragem) usada para lidar com conjuntos de dados desbalanceados, onde a classe minoritária tem poucas observações em comparação com a classe majoritária. O SMOTE gera novos exemplos sintéticos da classe minoritária, a partir dos exemplos existentes, para criar um conjunto de dados mais equilibrado.</p>

<p align="justify">O SMOTE funciona da seguinte forma: para cada observação na classe minoritária, o SMOTE seleciona k vizinhos mais próximos (com base na distância euclidiana) e cria novas observações sintéticas entre a observação original e seus vizinhos, adicionando um valor aleatório ponderado de cada uma das variáveis independentes. As novas observações sintéticas geradas são adicionadas ao conjunto de dados para equilibrar a distribuição de classes.</p>

<p align="justify">A vantagem do SMOTE é que ele pode melhorar o desempenho do modelo de machine learning ao lidar com conjuntos de dados desbalanceados, sem precisar coletar mais dados para a classe minoritária.</p>

```python
# Instanciar o SMOTE
smote = SMOTE(sampling_strategy='minority', random_state=42)

# Aplicar o SMOTE nos dados de treinamento
X_resampled, y_resampled = smote.fit_resample(X, y)

# Dividir o dataset entre Treino e Teste
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, stratify=y_resampled, test_size=0.3, random_state=42)

# Instanciar e treinar um modelo de Regressão Logística
model = LogisticRegression()
model.fit(X_train, y_train)

# Fazer as previsões em cima dos dados de teste
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)

# Plotar a matriz de confusão
skplt.metrics.plot_confusion_matrix(y_test, y_pred, normalize=True)
plt.show()

# Imprimir o relatório de classificação
print("Relatório de Classificação:\n", classification_report(y_test, y_pred, digits=4))

# Imprimir a acurácia
print("Acurácia: {:.4f}\n".format(accuracy_score(y_test, y_pred)))

# Imprimir a área sob a curva (AUC)
print("Área sob a curva (AUC): {:.4f}".format(roc_auc_score(y_test, y_proba[:,1])))
```

<p align="justify">O modelo de Regressão Logística treinado após a aplicação da técnica de "oversampling SMOTE" obteve uma acurácia de 0,9479, o que é um valor bastante razoável. A área sob a curva (AUC) foi de 0,9891, o que indica que o modelo teve um bom desempenho em classificar as observações corretamente.</p>

<p align="justify">Comparado ao primeiro relatório, o segundo modelo obteve uma acurácia um pouco mais baixa, mas uma área sob a curva (AUC) significativamente maior do que o primeiro modelo. Além disso, no segundo modelo, tanto a precision quanto o recall para a classe minoritária (classe 1) foram superiores a 0,9, o que indica que o modelo teve um bom desempenho em identificar corretamente a classe minoritária, sem sacrificar muito a precisão. Além disso, o valor F1-score para a classe minoritária foi de 0,9464, o que indica um bom equilíbrio entre precision e recall.</p>

<p align="justify">O melhor indicador do segundo relatório para detectar fraudes em cartão é o recall para a classe minoritária (classe 1), que obteve um valor de 0,9194. Isso significa que o modelo foi capaz de identificar corretamente 92% das transações fraudulentas. É importante destacar que, em problemas de detecção de fraudes em cartão, é mais importante minimizar os falsos negativos (transações fraudulentas classificadas como não fraudulentas) do que os falsos positivos (transações não fraudulentas classificadas como fraudulentas), pois as consequências de não identificar uma transação fraudulenta podem ser graves. Por esse motivo, o recall é um indicador mais relevante do que a precision para esse tipo de problema.</p>

<p align="justify">Em resumo, o modelo de Regressão Logística treinado após a aplicação do SMOTE obteve um desempenho bastante satisfatório, sendo capaz de lidar com o desbalanceamento do conjunto de dados e identificar corretamente a classe minoritária.</p>
